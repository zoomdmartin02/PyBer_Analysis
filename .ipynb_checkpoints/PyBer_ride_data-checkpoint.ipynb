```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics

# load in CSV
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]

# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar', color = "#87CEEB")
stdev = statistics.stdev(pyber_ride_df["Avg. Fare ($USD)"])
x_ax = pyber_ride_df["Month"]
y_ax = pyber_ride_df["Avg. Fare ($USD)"]
plt.errorbar(x_ax, y_ax, yerr=stdev, capsize=3)
plt.xticks(rotation=0)
plt.yticks(np.arange(0, 51, 5.0))
#plt.xticks(np.arange(min(x), max(x)+1, 1.0))
plt.show()
```
