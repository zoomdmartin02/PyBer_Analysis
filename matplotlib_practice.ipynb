```python
%matplotlib inline
```

```python
#Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, "-gD", linewidth=3)
# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add the legend.
ax.legend(["Boston"])
# Add grid lines
ax.grid()
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create horizontal bar chart
plt.barh(x_axis, y_axis)
```

```python
# Create horizontal bar chart
plt.barh(y_axis, x_axis)
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color="magenta")
plt.gca().invert_yaxis()
plt.legend(["Boston"])
plt.title("PyBer Fare by Month")
plt.ylabel("Date")
plt.xlabel("Fare($)")
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan")
ax.legend(["Chicago"])
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
# Create a title.
ax.set_title("PyBer Fare by Month")
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*6)
plt.scatter(y_axis, x_axis, s=y_axis_larger, color="r")
plt.legend(["Chicago"])
plt.title("PyBer Fare by Month")
plt.xlim(0,30)
plt.gca().invert_yaxis()
```

```python
plt.scatter(y_axis, x_axis, s= [i *3 for i in y_axis], color="r")
plt.legend(["Chicago"])
plt.title("PyBer Fare by Month")
plt.xlim(0,30)
plt.gca().invert_yaxis()
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color = "#87CEEB", s= [i * 5 for i in y_axis], alpha = .2, linewidth = 2, edgecolors = '#000000')
ax.set_xlim(0,30)
ax.invert_yaxis()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
explode_values = (0, 0, 0.3, 0, 0, 0, 0.3, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax = plt.subplots(figsize = (8, 8))
ax.pie(y_axis,labels=x_axis, explode=explode_values, autopct='%.1f%%', shadow = True, startangle=95, counterclock = False, colors=colors)


plt.show()
```
